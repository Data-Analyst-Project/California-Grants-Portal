### Cleaning DATA
* #### Lets drop Unwanted column.
* #### Status column to defoult index column.
* #### dop row in which status is closed.

```python
raw_data= raw_dataset.drop(['GrantID', 'AwardStats', 'ChangeNotes', 'LOI', 'Type', 'FundingSource', 'CategorySuggestion', 'FundingSourceNotes', 'OpenDate',
                            'MatchingFunds', 'GrantURL', 'ElecSubmission', 'ApplicantType', 'Description', 'Geography', 'EstAvailFunds', 'EstAwards', 'EstAmounts',
                            'FundingSource', 'MatchingFundsNotes', 'ApplicantTypeNotes', 'FundingMethodNotes', 'LastUpdated', 'AwardPeriod', 'ExpAwardDate', 'AgencySubscribeURL',  'GrantEventsURL', ], axis=1)
raw_data.set_index(['Status'], inplace = True)
raw_data.drop(labels = ["closed"], axis =0, inplace = True)
raw_data
```
